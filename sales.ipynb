Q1: Write the standard command to import the pandas library and load the file Video Games Sales (1980-2024) - Raw.csv into a variable named df.

```python
import pandas as pd
import numpy as np

url = 'videogame.csv'

df = pd.read_csv(url, sep= ',', engine='python', parse_dates=['release_date', 'last_update'])
df.head()
```

` Write the command that provides a technical summary of the DataFrame, including the number of non-null entries and the data type of every column.`

```python
df.info()
```

`Converting last_update and release_date to datetime format`

```python
df['release_date'] = pd.to_datetime(df['release_date'], dayfirst=True, errors = 'coerce')
df['last_update'] = pd.to_datetime(df['last_update'], dayfirst=True, errors = 'coerce')
```

#####  `🧹 Section 2: Dealing with Sparse Data (NaNs) AND clean up`

 ` Calculate the percentage of missing values in each column of the DataFrame. Which column has the highest percentage of NaNs?`

```python
## check % of missing values in each column

highest_missing = df.isnull().mean() * 100
highest_missing.sort_values(ascending=False)
```

```python
## fill up using  global median critic score
global_median = df['critic_score'].median()
df['critic_score'] = df['critic_score'].fillna(global_median)
```

```python
## fill up using median for the total sales column
total_sales_median = df['total_sales'].median()
df['total_sales'] = df['total_sales'].fillna(total_sales_median)
```

```python
## Replace the Unknown string in Publisher column to NaN to be recognized as missing
df['publisher'] = df['publisher'].replace(r'(?i)unknown', np.nan, regex=True)
```

```python
## upper case all the titles
df['title'] = df['title'].str.upper()
df['title'].head()
```

 `Duplicates, cleaning & removing redudancy`

```python
## check duplicated
duplicate_count = df.duplicated(subset=['title', 'console', 'release_date']).sum()

print(f"Number of duplicate records: {duplicate_count}")
```

```python
# Drop duplicates and update the DataFrame
df_clean = df.drop_duplicates(subset=['title', 'console', 'release_date'], keep='first')
```

```python
## remove the img column since its irrelevant
df_clean.drop('img', axis=1, inplace=True)
```

```python
##  removing the zero values in total sales

df_clean = df_clean[df_clean['total_sales'] > 0]

## check min value of total_sales
print(f'The min Total sales value is : {df_clean['total_sales'].min()}')
```

```python
## stripping white spaces 
df_clean['developer'] = df_clean['developer'].str.strip()
```

```python
df_clean.info()
```

`Memory optimization`

```python
# Shows the memory usage of each column in bytes before optimizing
df_clean.memory_usage(deep=True)
```

```python
## converting the genre and console into category datatype

category_columns = ['genre', 'console']

df_clean[category_columns] = df_clean[category_columns].astype('category')
```

```python
## resetting index to free some memory
df_clean.reset_index(drop=True, inplace=True)
```

```python
# Shows the memory usage of each column in bytes after optimizing
df_clean.memory_usage(deep=True)
```

```python
## Final memory usage and data types after cleaning
df_clean.info()
```

```python
## save the cleaned data to a new CSV file
df_clean.to_csv('videogame_cleaned.csv', index=False, encoding='utf-8')
```
