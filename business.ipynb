## 💰⚙️Implementing business logic based on the video games sales dataset

```python
import pandas as pd
import numpy as np
```

```python
### Loading Data
path = 'videogame_cleaned.csv'
df = pd.read_csv(path, sep= ',', parse_dates=['release_date','last_update'])
df.head()
```

### I. Portfolio & Publisher Strategy

```python
### The "Hit Maker" Ratio: Which publisher has the highest percentage of their library reaching "Hit" status (defined as total_sales > 5 million)?
hit_sales = df.groupby('publisher')['total_sales'].sum()
hit_sales.sort_values(ascending=False)
```
