## 💰⚙️Implementing business logic based on the video games sales dataset

```python
import pandas as pd
import numpy as np
```

```python
### Loading Data
path = 'videogame_cleaned.csv'
df = pd.read_csv(path, sep= ',', parse_dates=['release_date','last_update'])
df.head()
```

### I. Portfolio & Publisher Strategy

```python
### 1. The "Hit Maker" : Which publishers have the most individual 'Hit' titles (games with over 5 million in total sales) in their catalog?
hit_makers = df.query('total_sales > 5.0')
top_hit_makers = hit_makers.groupby('publisher')['total_sales'].count().sort_values(ascending=False)
print(f'Top Five hit makers pulishers {top_hit_makers.head(5)}')
```

```python
### 2. Publisher Efficiency: Portfolio Performance - Top Publishers by Per-Game Revenue

publisher_report = df.groupby('publisher').agg(
    avg_revenue=('total_sales', 'mean'),
    game_count=('title', 'count'),
    top_score=('critic_score', 'max')
)

publisher_reports = publisher_report.query('game_count > 5').nlargest(5, 'avg_revenue')
print(publisher_reports)
```

### Genre Diversification

```python
### Top 10 Most Prolific Publisher-Genre Pairings

publisher_games = df.groupby(['publisher', 'genre']).agg(
    no_of_games = ('genre', 'count')
)

publisher_lead_games = publisher_games.nlargest(10, 'no_of_games' , keep= 'first')
print(publisher_lead_games)
```

```python
### Top 10 Most Diversified Publishers by Genre Coverage

publisher_genre_counts = df.groupby('publisher')['genre'].nunique().nlargest(10)
publisher_genre_counts
```
